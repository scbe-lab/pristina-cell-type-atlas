# Declaring the input and output files

```python
adata = sc.read_h5ad('../../pristina_atlas_coloured_subcl.h5ad')
```

```python
adata
```

```python
adata.var
```

```python
clusteringlayer = 'leiden_1.5'
```

```python
list(adata.uns)
```

```python
with plt.rc_context({'figure.figsize': (12, 12)}):
    sc.pl.umap(adata, color=clusteringlayer, legend_loc='on data', legend_fontsize = 10, legend_fontoutline = 3, size = 10,
        frameon=False, add_outline = False,  save = '_'+figure+'_umapbig.pdf')
```

```python
adata.obs['broad_names'].cat.categories
```

```python
for i in adata.obs['broad_names'].cat.categories:
    with plt.rc_context({'figure.figsize': (12, 12)}):
        sc.pl.umap(adata, color=clusteringlayer, groups = adata.obs[adata.obs['broad_names'] == i][clusteringlayer].unique(),
            legend_loc='on data', na_color = 'whitesmoke', na_in_legend = False,
            legend_fontsize = 10, legend_fontoutline = 3, size = 10, title = i,
            frameon=False, add_outline = False, save = '_'+figure+'_'+i+'.pdf')
```

```python
 with plt.rc_context({'figure.figsize': (12, 12)}):
        sc.pl.umap(adata, color=clusteringlayer, groups = adata.obs[adata.obs['broad_names'] == 'unnanotated'][clusteringlayer].unique(),
            na_color = 'whitesmoke', na_in_legend = False,
            legend_fontsize = 10, legend_fontoutline = 3, size = 20, title = 'unnanotated highligted',
            palette = sns.color_palette("Set2", 60),
            frameon=False, add_outline = False, save = '_'+figure+'_'+'unnanotated_highlighted'+'.pdf')
```
